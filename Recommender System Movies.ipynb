Item based collaborative technique

```python
# creating pivot table to create item by item collaborative filtering
movie_rating_pivot = pd.pivot_table(data=ratings, values='rating', index='userId', columns='movieId')
#movie_rating_pivot.fillna(0, inplace=True)
```

```python
movie_rating_pivot.head()
```

```python
top_popular_movieId =356
```

```python
FG_rating= movie_rating_pivot[top_popular_movieId] #fg is forest gump
FG_rating[FG_rating>=0]
```

```python
similar_to_FG = movie_rating_pivot.corrwith(FG_rating)
similar_to_FG 
```

```python
FG_correlation= pd.DataFrame(similar_to_FG, columns=['PearsonR'])
FG_correlation.dropna(inplace=True)
FG_correlation.head(10)
```

```python
rating = pd.DataFrame(ratings.groupby('movieId')['rating'].mean())
rating['rating_count'] =ratings.groupby('movieId')['rating'].count()
```

```python
FG_corr_summary = FG_correlation.join(rating['rating_count'])
FG_corr_summary.drop(top_popular_movieId, inplace=True) 
FG_corr_summary
```

Picking the top 10 movies

```python
top10 = FG_corr_summary[FG_corr_summary['rating_count']>=10].sort_values('PearsonR', ascending=False).head(10)
top10
```

```python
def top_n_movies(movieId, n):
    movie_ratings = movie_rating_pivot[movieId]
    similar_to_movies = movie_rating_pivot.corrwith(movie_ratings)
    corr_movies = pd.DataFrame(similar_to_movies, columns=['PearsonR'])
    corr_movies.dropna(inplace=True)
    movies_corr_summary = corr_movies.join(rating['rating_count'])
    movies_corr_summary.drop(movieId, inplace=True) # drop the inputed restaurant itself
    top10 = movies_corr_summary[movies_corr_summary['rating_count']>=10].sort_values('PearsonR', ascending=False).head(n)
    top10 = top10.merge(movies, left_index=True, right_on="movieId")
    return list(top10["title"])
```

User_based Collaborative technique

```python
# creating pivot table to create item by item collaborative filtering
user_pivot = pd.pivot_table(data=ratings, values='rating', index='userId', columns='movieId')
user_pivot.head()
```

```python
#replacing NaNs with zeros
user_pivot.fillna(0, inplace=True)
user_pivot.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarities = pd.DataFrame(cosine_similarity(user_pivot),
                                 columns=user_pivot.index, 
                                 index=user_pivot.index)
user_similarities.head()
```

```python
# compute the weights for one user
user_id = 1

weights = (
    user_similarities.query("userId!=@user_id")[user_id] / sum(user_similarities.query("userId!=@user_id")[user_id])
          )
weights.head(6)
```

```python
weights.sum()
```

```python
user_pivot.loc[user_id,:]==0
```

```python
# select movies that the inputed user has not been rated
not_rated_movies = user_pivot.loc[user_pivot.index!=user_id, user_pivot.loc[user_id,:]==0]
not_rated_movies.T
```

```python
# dot product between the not-visited-restaurants and the weights
weighted_averages = pd.DataFrame(not_rated_movies.T.dot(weights), columns=["predicted_rating"])
weighted_averages.head()
```

```python
recommendations = weighted_averages.merge(movies, left_index=True, right_on="movieId")
recommendations.sort_values("predicted_rating", ascending=False).head()
```

```python
def top_n_movies_user(userId, n):
   user_similarities = pd.DataFrame(cosine_similarity(user_pivot), columns=user_pivot.index, index=user_pivot.index)  
   weights = (user_similarities.query("userId!=@user_id")[user_id] / sum(user_similarities.query("userId!=@user_id")[user_id]))  
   not_rated_movies = user_pivot.loc[user_pivot.index!=user_id, user_pivot.loc[user_id,:]==0]
   weighted_averages = pd.DataFrame(not_rated_movies.T.dot(weights), columns=["predicted_rating"])
   recommendations = weighted_averages.merge(movies, left_index=True, right_on="movieId")
   top_movies=recommendations.sort_values("predicted_rating", ascending=False).head()
   return top_movies
```
